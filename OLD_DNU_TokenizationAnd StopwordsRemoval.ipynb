```python
import pandas as pd
import numpy as np
import nltk
import string  # Importing the string module for punctuation
import re  
from nltk.corpus import stopwords
from collections import Counter

file_path = r'D:\data_trustpilot.csv'
df = pd.read_csv(r'D:\data_trustpilot.csv')
print(df.head(5))
df.info()

# Selecting the relevant columns for WordCloud
wordcloud_df = df[['text', 'subject', 'answer']]
print('\n')
wordcloud_df.info()
print(wordcloud_df.head())

# Removing the stop words
stop_words = set(stopwords.words('english'))

# Function to clean the text by removing stopwords, punctuation, and special characters.
def clean_text(text):
    if isinstance(text, str):  
        # Remove special characters and digits using regex
        text = re.sub(r'[^A-Za-z\s]', '', text)
        
        # Split the text into words
        words = text.split()         
        
        # Removing stopwords and convert to lowercase
        filtered_words = [word.lower() for word in words if word.lower() not in stop_words]
        
        return ' '.join(filtered_words)  
    return text  

# Applying the clean_text function to the text, subject and answer columns in the dataframe.
wordcloud_df['text'] = wordcloud_df['text'].apply(clean_text)
wordcloud_df['subject'] = wordcloud_df['subject'].apply(clean_text)
wordcloud_df['answer'] = wordcloud_df['answer'].apply(clean_text)

# Displaying the updated DataFrame
print(wordcloud_df.head())

# Calculating the average length of text in the 'text' column
wordcloud_df['text_length'] = wordcloud_df['text'].apply(len)  
average_length = wordcloud_df['text_length'].mean()  # Calculating the average length
print(f"Average length of text: {average_length:.2f} characters")

# Calculating the average length of text in the 'subject' column
wordcloud_df['subject_length'] = wordcloud_df['subject'].apply(len)
average_subject_length = wordcloud_df['subject_length'].mean()  # Calculating the average length for 'subject'
print(f"Average length of subject: {average_subject_length:.2f} characters")


# Combining all the text and subject into a single string for common words analysis
all_text = ' '.join(wordcloud_df['text']) + ' ' + ' '.join(wordcloud_df['subject'])

# Tokenizing the combined text and converting to lowercase for counting
words = all_text.lower().split()

# Counting the frequency of each word
word_counts = Counter(words)

# Getting the 500 most common words
common_words = word_counts.most_common(500)

# Displaying the common words
print("Most common words:")
for word, count in common_words:
    print(f"{word}: {count}")
```
