```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords

from collections import Counter 


file_path = 'data/data_trustpilot.csv'
df = pd.read_csv(file_path)
print(df.head(5))
df.info()

# Selecting the relevant columns for WordCloud
wordcloud_df = df[['text', 'subject', 'answer']]
print('\n')
wordcloud_df.info()
print(wordcloud_df.head())

#Removing the stop words
stop_words = set(stopwords.words('english'))

# Function to remove stop words from a given text
def remove_stopwords(text):
    if isinstance(text, str): 
        words = text.split() 
        filtered_words = [word for word in words if word.lower() not in stop_words]
        return ' '.join(filtered_words)  
    return text 

# Applying the remove_stopwords function to the relevant columns in the dataframe.
wordcloud_df['text'] = wordcloud_df['text'].apply(remove_stopwords)
wordcloud_df['subject'] = wordcloud_df['subject'].apply(remove_stopwords)
wordcloud_df['answer'] = wordcloud_df['answer'].apply(remove_stopwords)

# Displaying the updated DataFrame
print(wordcloud_df.head())

# Calculate average length of text in the 'text' column
wordcloud_df['text_length'] = wordcloud_df['text'].apply(len)  
average_length = wordcloud_df['text_length'].mean()  # Calculating the average length
print(f"Average length of text: {average_length:.2f} characters")

# Combining all the text into a single string for common words analysis
all_text = ' '.join(wordcloud_df['text'])

# Tokenizing the text and convert to lower case for counting
words = all_text.lower().split()

# Counting the frequency of each word
word_counts = Counter(words)

# Get the 50 most common words
common_words = word_counts.most_common(50)

# Displaying the common words
print("Most common words:")
for word, count in common_words:
    print(f"{word}: {count}")

```
